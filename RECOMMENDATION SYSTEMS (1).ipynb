```python
import numpy as np
import pandas as pd

def compute_metrics_at_5(group):
    actual_cluster = group["actual_cluster"].iloc[0]

    # Keep top 5
    group = group.nsmallest(5, "rank")  # ensure ranked from 1 to 5
    correct_mask = (group["cluster_selected"] == actual_cluster).astype(int)
    
    # Accuracy@5: at least one relevant
    accuracy = int(correct_mask.any())
    
    # Precision@5: how many are relevant
    precision = correct_mask.sum() / 5

    # MRR@5: reciprocal of first correct rank
    if correct_mask.any():
        first_correct_rank = group[correct_mask == 1]["rank"].min()
        mrr = 1 / first_correct_rank
    else:
        mrr = 0.0

    # NDCG@5
    dcg = (correct_mask / np.log2(group["rank"] + 1)).sum()
    ideal_dcg = sum(1.0 / np.log2(i + 2) for i in range(correct_mask.sum()))  # ideal case
    ndcg = dcg / ideal_dcg if ideal_dcg > 0 else 0.0

    return pd.Series({
        "accuracy_at_5": accuracy,
        "precision_at_5": precision,
        "mrr_at_5": mrr,
        "ndcg_at_5": ndcg
    })
```

```python

metrics_df_lyrics_only = (
    results_df_lyrics_only
    .groupby(["column", "image_name"])
    .apply(compute_metrics_at_5)
    .reset_index()
)
```

```python
metrics_df = (
    results_df_new_1
    .groupby(["column", "image_name"])
    .apply(compute_metrics_at_5)
    .reset_index()
)
```
