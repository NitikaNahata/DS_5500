```python
def get_all_null_columns(df):
    """
    Returns a list of column names that contain only null values in the DataFrame.
    
    Parameters:
    df (pandas.DataFrame): The DataFrame to check
    
    Returns:
    list: Column names that have only null values
    """
    # Check which columns have all null values
    all_null_mask = df.isna().all()
    
    # Get the column names where all values are null
    all_null_columns = all_null_mask[all_null_mask].index.tolist()
    
    return all_null_columns
```

```python
# Get columns with all null values
null_columns = get_all_null_columns(df_inference)
print("Columns with all null values:", null_columns)
```

```python
def remove_all_null_columns(df):
    """
    Removes all columns that contain only null values from the DataFrame.
    
    Parameters:
    df (pandas.DataFrame): The DataFrame to process
    
    Returns:
    pandas.DataFrame: A new DataFrame with all-null columns removed
    """
    # Get a boolean mask where True means the column has all null values
    all_null_mask = df.isna().all()
    
    # Get column names that don't have all null values
    columns_to_keep = all_null_mask[~all_null_mask].index.tolist()
    
    # Return a new DataFrame with only the columns to keep
    return df[columns_to_keep]
```

```python
def refactor_column_names_programmatically(df, columns_to_refactor):
    """
    Refactors specified column names by adding "Reasoning_" prefix.
    
    Parameters:
    df (pandas.DataFrame): The DataFrame with columns to rename
    columns_to_refactor (list): List of column names to be refactored
    
    Returns:
    pandas.DataFrame: DataFrame with renamed columns
    """
    # Create the renaming dictionary programmatically
    columns_to_rename = {col: f"Reasoning_{col}" for col in columns_to_refactor}
    
    # Rename the columns in the DataFrame
    df = df.rename(columns=columns_to_rename)
    
    return df
```
