```python
import requests
import pandas as pd
import time

# Load your dataset
df = pd.read_csv("hindi_only.csv")

# OpenRouter API setup
API_KEY = ""  # Replace with your actual API key
API_URL = "https://openrouter.ai/api/v1/chat/completions"
HEADERS = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

# Function to summarize lyrics
def summarize_lyrics(lyrics):
    prompt = f"""
        These are the lyrics of a song. Summarize the song in 2 sentences in English, capturing the emotional tone, theme, and overall mood. 
        The lyrics may be in Hindi or English. If in Hindi, still provide the summary in English.

        Lyrics:
        {lyrics}
        """

    payload = {
        "model": "mistralai/Mixtral-8x7B-Instruct-v0.1",  # Specify the Mixtral model
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.7,
        "max_tokens": 100
    }
    try:
        response = requests.post(API_URL, headers=HEADERS, json=payload)
        response.raise_for_status()
        summary = response.json()["choices"][0]["message"]["content"].strip()
        return summary
    except Exception as e:
        print(f"Error summarizing lyrics: {e}")
        return None

# Apply summarization to the first 10 rows for testing
df_sample = df.head(10).copy()
df_sample["summary"] = df_sample["lyrics"].apply(lambda x: summarize_lyrics(x) if pd.notnull(x) else None)

# Save the results
df_sample.to_csv("sample_summaries_hindii.csv", index=False)
print("Sample summaries saved to 'sample_summaries.csv'")
```

```python
import requests
import pandas as pd
import time
import re


class LyricsSummarizer:
    def __init__(self, api_key, model="mistralai/Mixtral-8x7B-Instruct-v0.1", api_url="https://openrouter.ai/api/v1/chat/completions"):
        self.api_key = api_key
        self.model = model
        self.api_url = api_url
        self.headers = {
            "Authorization": f"Bearer {self.api_key}",
            "Content-Type": "application/json"
        }

    def clean_lyrics(self, lyrics):
        """Remove section headers like [Chorus], [Verse] etc."""
        return re.sub(r"\[.*?\]", "", lyrics).strip()

    def summarize_lyrics(self, lyrics):
        """Send lyrics to the Mixtral API and get back a summary."""
        prompt = f"""
        These are the lyrics of a song. Summarize the song in 2 sentences in English, capturing the emotional tone, theme, and overall mood. 
        The lyrics may be in Hindi or English. If in Hindi, still provide the summary in English.

        Lyrics:
        {lyrics}
        """

        payload = {
            "model": self.model,
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.7,
            "max_tokens": 100
        }

        try:
            response = requests.post(self.api_url, headers=self.headers, json=payload)
            response.raise_for_status()
            summary = response.json()["choices"][0]["message"]["content"].strip()
            return summary
        except Exception as e:
            print(f"⚠️ Error summarizing lyrics: {e}")
            return None

    def process_dataframe(self, df, output_csv=None, start_idx=0, autosave_every=50):
        """Apply summarization to a DataFrame of songs."""
        if "summary" in df.columns:
            summaries = df["summary"].tolist()
        else:
            summaries = [None] * len(df)

        for idx in range(start_idx, len(df)):
            lyrics = df.iloc[idx].get("lyrics")

            if pd.notnull(lyrics) and (not summaries[idx] or pd.isna(summaries[idx])):

                print(f"Summarizing {idx + 1}/{len(df)}")
                summaries[idx] = self.summarize_lyrics(lyrics)
                time.sleep(1.2)

            # Save progress periodically
            if output_csv and (idx + 1) % autosave_every == 0:
                df["summary"] = summaries
                df.to_csv(output_csv, index=False)
                print(f"Autosaved at row {idx + 1} → {output_csv}")

        # Final save
        df["summary"] = summaries
        if output_csv:
            df.to_csv(output_csv, index=False)
            print(f"Final save completed → {output_csv}")

        return df
```
